### **Broadcasting Rule 2**

```python
A = np.array([1, 2, 3])
# Shape (3,)
B = np.array([[10], [20], [30]]) # Shape (3,1)
print(A+B, 'A+B')
print(A*B, 'A*B')
```

### **Broadcasting Rule 3**

 Rule 3: If any dimension is incompatible (neither same nor 1), broadcasting fails
 

```python
A = np.array([[1, 2, 3], [4, 5, 6]]) # Shape (2,3)
print(A)
B = np.array([10, 20]) # Shape (2,)
print(B)
try:
    print(A + B) # Incompatible as 3 doesn't equal 2
except ValueError as e:
    print("Broadcasting error:", e)
```

```python
A = np.array([[1, 2, 3], [4, 5, 6]])
print(A, 'A')
B = np.array([10, 20])
B = np.expand_dims(B, axis=1)
print(B, 'B')
print(A+B, 'A+B')
```

```python
x = np.array([1, 2, 3]) # Shape (3,)
X = np.expand_dims(x, axis=0) # Shape (1,3)
print(X, 'X', X.shape)
print(x, 'x', x.shape)
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2*math.pi,100)
y = np.sin(x)
```

```python
ax = plt.gca(); # 'Get current axis'
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('$sin(x)$'); #+
ax.set_title('Plot of the Sine Function')
```
