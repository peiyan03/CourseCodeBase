Suppose `x` is a list of integers.

2. Write code that produces a list `z` containing only the positive integers of `x`.

````python
# Define an example list of integers:
x = [-10, 2, 30, -4]

# Part 1:

y = []
for i in x:
    y.append(i)
print("Printing list y:")
print(y)

# Part 2:
z = []
for i in x:
    if i > 0:  # 0 is not positive, so we don't want >=
        z.append(i)
print("Printing list z containing only the positive integers of x:")
print(z)
````

## Exercise 7

The Caesar cipher works by shifting each character in the original message (plaintext) by a fixed number of positions up. 

For example, with a shift of 3, 'A' becomes 'D', 'B' becomes 'E', and so on. And the message "HELLO" becomes ""

Implement the Caesar cipher encryption.

**Hints:**
1. Define three variables, `plaintext` and `shift` and `ciphertext`. What types of values should these be or contain? `ciphertext` should be empty, initially.
2. Recall from the lecture that it's possible in Python to convert characters from and to (numerical) ASCII values using the `chr()` and `ord()` functions.
3. Use a for loop to iterate over each character in `plaintext`, shift the character, and append the shifted character to the `ciphertext`.

````python
plaintext = "Hello World"
shift = 3
ciphertext = ""
for char in plaintext:
    ascii_num = ord(char)
    shifted_num = ascii_num + shift
    shifted_char = chr(shifted_num)
    ciphertext = ciphertext + shifted_char
print(ciphertext)
````

## Exercise 9

Write code that converts temperature `x` from Fahrenheit to Celsius, and temperature `y` from Celsius to Fahrenheit.

From Wikipedia:

$$
x \degree F = \frac{5}{9} \times (x-32) \degree C.
$$

$$
y \degree C = (y \times \frac{5}{9} + 32) \degree F.
$$

````python
# Converting from Fahrenheit to Celsius
f = 99
c = 5 / 9 * (f - 32)
print("Temperature " + str(f) + "˚F converts to " + str(c) + "˚C.")

# Converting from Celsius to Fahrenheit
c = 37
f = c * 9 / 5 + 32
print("Temperature " + str(c) + "˚C converts to " + str(f) + "˚F.")
````
